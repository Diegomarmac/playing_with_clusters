```python
from sklearn.cluster import SpectralClustering, KMeans
from sklearn.datasets import make_blobs
from sklearn.metrics import pairwise_distances

from sklearn.datasets import make_circles
from numpy import random

import matplotlib.pyplot as plt
import numpy as np
import networkx as nx
import seaborn as sns
```

```python
data, clusters = make_circles(n_samples=2000, noise=0.1, factor=0.3, random_state=5)
plt.scatter(data[:,0], data[:,1])
```

```python
kmeans = KMeans(init='k-means++', n_clusters=2)
km_clustering = kmeans.fit(data)
plt.scatter(data[:,0],data[:,1], c=km_clustering.labels_, cmap='prism', alpha=0.8)
```

Se observa que con K-means a dos grupos, no se discrimina adecuadamente, parte a la mitad el conjunto de datos. lo que queremos es que cada circulo sea un cluster... !Spectral Clustering!

```python
spectral = SpectralClustering(n_clusters=2, affinity="nearest_neighbors", random_state=5)
sc_clustering = spectral.fit(data)

plt.scatter(data[:,0], data[:,1], c=sc_clustering.labels_, cmap='prism', alpha=0.8)
```

Usando Spectral Cluster se observa como cada circulo pertenece a un cluster distinto, que es lo que esperabamos !
